```python
# Imports
import torch
from transformers import pipeline
from diffusers import DiffusionPipeline
from datasets import load_dataset
import soundfile as sf
from IPython.display import Audio
```

```python
# Sentiment analysis
classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life).")
```

```python
# Named Entity Recognition
ner = pipeline("ner")
result = ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
print(result)
```

```python
# Question-answering
question_answer = pipeline('question-answering')
result = question_answer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn"
)
print(result)
```

```python
# Text summerization
summerizer = pipeline("summarization")
result = summerizer("""
My name is anuj kumar. I am currently working in a f2 fintech company as Ml intern,
""")
print(result)
```

```python
# Tranlator
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
result = translator("Ce cours est produit par Hugging Face.")
print(result)
```
